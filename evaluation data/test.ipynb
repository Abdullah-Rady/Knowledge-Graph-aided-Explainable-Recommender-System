```python
ratings_df = pd.read_csv(f'{path}/../ml1m/preprocessed/ratings.txt', sep='\t')
ratings_df
```

```python
movies_df = pd.read_csv(f'{path}/../ml1m/movies.dat',  sep="::", names=["movie_id", "movie_name", "genre"], header=None, encoding='latin-1')
movies_df
```

```python
mapping_df = pd.read_csv(f'{path}/../ml1m/preprocessed/products.txt', sep='\t', index_col='new_id')
mapping_df
```

```python
# Load the data
movies = pd.merge(movies_df, mapping_df, left_on='movie_id', right_on='raw_dataset_id')
movies
```

```python
movies['id'] = range(len(movies))
movies.set_index('id', inplace=True)
movies = movies.drop(columns=['movie_id', 'raw_dataset_id'])
movies
```

```python
data = pd.merge(ratings_df, movies, left_on='pid', right_on='id')
data
```

```python
def final(df):
    max_userid = max(data['uid'])
    df['uid'] = max_userid + 1
    a = pd.merge(df, movies_df, left_on='raw_dataset_id', right_on='movie_id')
    df_final = a.drop(columns=['raw_dataset_id', 'movie_id', 'movie_name_x', 'entity_id', ])
    df_final = df_final.rename(columns={'movie_name_y': 'movie_name'})
    df_final = df_final[['uid','pid', 'rating', 'movie_name', 'genre']]
    return df_final
```

```python
r1_final = final(r1)
data = data.append(r1_final)

r2_final = final(r2)
data = data.append(r2_final)

r3_final = final(r3)
data = data.append(r3_final)

r4_final = final(r4)
data = data.append(r4_final)

r5_final = final(r5)
data = data.append(r5_final)

r6_final = final(r6)
data = data.append(r6_final)

r7_final = final(r7)
data = data.append(r7_final)

r8_final = final(r8)
data = data.append(r8_final)

r9_final = final(r9)
data = data.append(r9_final)

r11_final = final(r11)
data = data.append(r11_final)

r12_final = final(r12)
data = data.append(r12_final)
```

```python
user_movie_matrix = data.pivot_table(index='uid', columns='movie_name', values='rating').fillna(0)
user_movie_matrix
```

```python
user_movie_matrix_norm = user_movie_matrix.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)), axis=1).clip(0)
user_movie_matrix_norm
```

```python
sparse_matrix = csr_matrix(user_movie_matrix.values)
sparse_matrix
```

```python
similarities = cosine_similarity(sparse_matrix)
similarities
```

```python
collab_filtering = pd.DataFrame(similarities, index=user_movie_matrix.index, columns=user_movie_matrix.index)
collab_filtering
```

```python
movie_features = movies.genre.str.get_dummies()
movie_features
```

```python
movie_similarities = cosine_similarity(movie_features)
movie_similarities
```

```python
content_based = pd.DataFrame(movie_similarities, index=movies['movie_name'], columns=movies['movie_name'])
content_based
```

```python
def get_recommendation(user_id):
    # Get top 5 similar users
    similar_users = collab_filtering[user_id].sort_values(ascending=False)[1:6].index.tolist()
    similar_users
    # Get movies rated highly by similar users
    similar_user_movies = user_movie_matrix.loc[similar_users].mean(axis=0)
    recommended_movies = similar_user_movies[similar_user_movies >= 2].index.tolist()
    recommended_movies
    user_movies = user_movie_matrix.loc[user_id]
    similar_movies = content_based[user_movies > 0].mean(axis=0).sort_values(ascending=False)[:5].index.tolist()
    similar_movies
    recommendations = set(recommended_movies) | set(similar_movies)
    recommendations

    rec = []

    for i in recommendations:
        if len(r1_final[r1_final['movie_name'] == i]) == 0:
            rec.append(i)

    return rec
```
